## Import Libraries

*Import libraries and set up plotting style.*

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

## Load Image

*Load the original image and convert it to grayscale for processing.*

```python
# Load and convert image
img = cv2.imread("flower.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

## Image Filtering (Blurring & Edge Detection)

*Visualize how Gaussian blur smooths an image and how Canny edge detection highlights edges.*

```python
# Apply Gaussian blur
blurred = cv2.GaussianBlur(gray, (25,25), 0)  # Very strong blur

# Apply Canny edge detection
edges = cv2.Canny(gray, 50, 150)

# Plot results
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(blurred, cmap='gray')
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")
plt.axis("off")

plt.tight_layout()
plt.savefig("filtering_edges.png", bbox_inches="tight", pad_inches=0.1)
plt.show()
```

## Thresholding (Black-and-White Conversion)

*Convert a grayscale image to black-and-white using thresholding. Pixels brighter than the threshold become white, others black.*


```python
# Apply simple binary threshold
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(6,6))
plt.imshow(thresh, cmap='gray')
plt.title("Binary Thresholding")
plt.axis("off")
plt.savefig("threshold.png", bbox_inches="tight", pad_inches=0.1)
plt.show()
```

## Contour Detection

*Detect and highlight the contours of objects in the image based on the thresholded black-and-white regions.*

```python
# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the image
img_contours = img_rgb.copy()
cv2.drawContours(img_contours, contours, -1, (255,0,0), 2)

plt.figure(figsize=(6,6))
plt.imshow(img_contours)
plt.title("Contours Detected")
plt.axis("off")
plt.savefig("contours.png", bbox_inches="tight", pad_inches=0.1)
plt.show()
```
